```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load data
data = pd.read_csv('AirPassengers.csv')
# Convert Month to numbers (like 0, 1, 2, ...)
data['Month_Num'] = range(len(data))
print(data)
# Prepare features and target
x = data[['Month_Num']]
y = data['Passengers']

# Train model
model = LinearRegression()
model.fit(x, y)

# Predict
predictions = model.predict(x)

# Plot + scatter
plt.scatter(x,y, color = 'grey', label='Data')
plt.plot(y, label='Actual')
plt.plot(predictions, label='Predicted', linestyle='--', color='red')
plt.xlabel('Time (Months)')
plt.ylabel('Passengers')
plt.title('Air Passengers Prediction')
plt.legend()
plt.show()
```

```python
#plot
plt.scatter(x, y)
plt.plot(x, model.predict(x), color='blue')
plt.xlabel('Experience')
plt.ylabel('Salary')
plt.title('Salary Predection with Experience')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
```

```python
#load data
data1 = {'Rooms': [1,2,3,4,5], 'Area':[300, 700, 1100, 1300, 1500], 'price_in_million':[3,5,8,10,14]}
df= pd.DataFrame(data1)
```

```python
x= df[['Rooms', 'Area']]
y= df['price_in_million']

x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=0)
model= LinearRegression()
```

```python
model.fit(x_train, y_train)
y_pred= model.predict(x_test)
print("MSE", mean_squared_error(y_test, y_pred))
```

```python
print("Price for n number of rooms and x sqft of areas", model.predict([[7, 2000]]))
```

```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Scatter plot of the data points
ax.scatter(df['Rooms'], df['Area'], df['price_in_million'], c='red', marker='o')

# Create a meshgrid for the regression plane
x_surf, y_surf = np.meshgrid(np.linspace(df['Rooms'].min(), df['Rooms'].max(), 10),
                             np.linspace(df['Area'].min(), df['Area'].max(), 10))

# Predict the prices for the meshgrid points
z_surf = model.predict(np.c_[x_surf.ravel(), y_surf.ravel()])
z_surf = z_surf.reshape(x_surf.shape)

# Plot the regression plane
ax.plot_surface(x_surf, y_surf, z_surf, color='blue', alpha=0.5)

ax.set_xlabel('Rooms')
ax.set_ylabel('Area')
ax.set_zlabel('Price in Million')
ax.set_title('Housing Price Prediction (3D)')

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
```

```python
data={'Room': [1,2,3,4,5,6], 'Price':[10,20,35,40,47,58]}
df= pd.DataFrame(data)
```

```python
x= df[['Room']]
y= df['Price']
```

```python
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.33, random_state=0)
```

```python
model= LinearRegression()
model.fit(x_train, y_train)
```

```python
y_pred= model.predict(x_test)
```

```python
plt.scatter(x_test, y_test, color='blue',label="Actual")
plt.plot(x_test, y_pred, color="red", linewidth=2, label='Predicted')
plt.xlabel("Number of rooms")
plt.ylabel("House price")
plt.legend()
plt.grid(True)
plt.show()
```

```python
mae= mean_absolute_error(y_test, y_pred)
mse= mean_squared_error(y_test, y_pred)
rmse= np.sqrt(mse)
r2= r2_score(y_test, y_pred)
print("Evluation:")
print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
print(f"R2: {r2:.2f}")
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Data
details = {
    'roll no ': [1, 2, 3, 4, 6, 7],
    'marks': [10, 20, 30, 40, 50, 60]
}

data = pd.DataFrame(details)
x = data[['roll no ']]
y = data[['marks']]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Train the model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict on test data
y_pred = model.predict(x_test)

# Evaluation metrics
print("MSE:", mean_squared_error(y_test, y_pred))
print("R Score:", r2_score(y_test, y_pred))

#  Plot all data
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.scatter(x_test, y_test, color='green', label='Test Data')
plt.plot(x_test, y_pred, color='grey', linewidth=2, label='Regression Line (Test Prediction)')

# Labels and legend
plt.xlabel("Student Roll No")
plt.ylabel("Marks")
plt.title("Student Marks Prediction")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, mean_squared_error

# Data
details = {
    'roll no ': [1, 2, 3, 4, 6, 7],
    'marks': [10, 20, 30, 40, 50, 60]
}

data = pd.DataFrame(details)
x = data[['roll no ']]
y = data[['marks']]

# Train-test split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=0)

# Train model
model = LinearRegression()
model.fit(x_train, y_train)

# Predict on full x for extended line
y_line = model.predict(x)

# Plot all data and regression line
plt.scatter(x_train, y_train, color='blue', label='Training Data')
plt.scatter(x_test, y_test, color='green', label='Test Data')
plt.plot(x, y_line, color='red', label='Regression Line')

plt.xlabel("Student Roll No")
plt.ylabel("Marks")
plt.title("Simple Extended Regression Line")
plt.legend()
plt.grid(True)
plt.show()
```
